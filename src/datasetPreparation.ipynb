```python
import pandas as pd

data=pd.read_csv('..\Dataset\\food.csv')

data = data.set_index(data.index)
data.index.name = "ID"

data.head()
```

```python
breakfast=data.copy(deep=True)
lunch=data.copy(deep=True)
dinner=data.copy(deep=True)
```

```python
breakfast.drop(breakfast.loc[breakfast['Breakfast']!=1].index, inplace=True,axis=0)
breakfast = breakfast.drop(['Food_items', 'Breakfast', 'Dinner', 'Lunch'],axis=1)
breakfast.head()
```

```python
lunch.drop(lunch.loc[lunch['Lunch']!=1].index, inplace=True,axis=0)
lunch = lunch.drop(['Food_items', 'Breakfast', 'Dinner', 'Lunch'],axis=1)
lunch.head()
```

```python
dinner.drop(dinner.loc[dinner['Dinner']!=1].index, inplace=True,axis=0)
dinner = dinner.drop(['Food_items', 'Breakfast', 'Dinner', 'Lunch'],axis=1)
dinner.head()
```

```python
breakfast.to_csv('..\\Dataset\\Processed_dataset\\breakfast.csv', index_label="ID")
lunch.to_csv('..\\Dataset\\Processed_dataset\\lunch.csv', index_label="ID")
dinner.to_csv('..\\Dataset\\Processed_dataset\\dinner.csv', index_label="ID")
```
